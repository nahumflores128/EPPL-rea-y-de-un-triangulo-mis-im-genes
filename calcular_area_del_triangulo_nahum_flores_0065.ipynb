## Formula para calcular el area
$$Area = \frac{base \times altura}{2}$$

---
## Datos de entrada
b = ?

```python
# calcular area de un triangulo
# variables b,h,
# formula area = (b*h) / 2
import math
b = float(input("Ingresa la base del triángulo: "))
h = float(input("Ingresa la altura del triángulo: "))
area = (b*h) / 2
print(f"El área del triángulo es: {area}")
```
